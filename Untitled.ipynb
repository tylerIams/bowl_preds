```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from scipy.stats import zscore
from sklearn.metrics import accuracy_score
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df2008 = pd.read_csv("game_data/game_dat_2008.csv")
df2009 = pd.read_csv("game_data/game_dat_2009.csv")
df2010 = pd.read_csv("game_data/game_dat_2010.csv")
df2011 = pd.read_csv("game_data/game_dat_2011.csv")
df2012 = pd.read_csv("game_data/game_dat_2012.csv")
df2013 = pd.read_csv("game_data/game_dat_2013.csv")
df2014 = pd.read_csv("game_data/game_dat_2014.csv")
df2015 = pd.read_csv("game_data/game_dat_2015.csv")
df2016 = pd.read_csv("game_data/game_dat_2016.csv")
df2017 = pd.read_csv("game_data/game_dat_2017.csv")
df2018 = pd.read_csv("game_data/game_dat_2018.csv")
```

```python
df2018_val = df2018.loc[df2018["Wk"] < 17]
df2018_test = df2018.loc[df2018["Wk"] >= 17]
```

```python
def scale_zscore(df):
    columns = df.columns.difference(["Wk", "Year", "Home", "Away", "Outcome"])
    df[columns] = df[columns].apply(zscore)
    df = df.dropna()
    return df[columns], df["Outcome"]
```

```python
x2008, y2008 = scale_zscore(df2008)
x2009, y2009 = scale_zscore(df2009)
x2010, y2010 = scale_zscore(df2010)
x2011, y2011 = scale_zscore(df2011)
x2012, y2012 = scale_zscore(df2012)
x2013, y2013 = scale_zscore(df2013)
x2014, y2014 = scale_zscore(df2014)
x2015, y2015 = scale_zscore(df2015)
x2016, y2016 = scale_zscore(df2016)
x2017, y2017 = scale_zscore(df2017)
x2018_val, y2018_val = scale_zscore(df2018_val)
x2018_test, y2018_test = scale_zscore(df2018_test)
```

```python
train_x = pd.concat([x2008, x2009, x2010, x2011, x2012, x2013, x2014, x2015, x2016, x2017])
train_y = pd.concat([y2008, y2009, y2010, y2011, y2012, y2013, y2014, y2015, y2016, y2017])
val_x = x2018_val
val_y = y2018_val
test_x = x2018_test
test_y = x2018_test
```

```python
clf = SVC(gamma='auto')
clf.fit(train_x, train_y)
```

```python
result = clf.predict(val_x)
result
```

```python
accuracy_score(val_y, result)
```
